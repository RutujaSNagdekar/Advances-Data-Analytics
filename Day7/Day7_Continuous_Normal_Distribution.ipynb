```python
#Continuous probability Distribution
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
df = pd.read_csv('hs_heights_pair.csv')
df
```

```python
boys_heights = df['boys']
girls_heights = df['girls']
```

```python
mean_boys = np.mean(boys_heights)
std_boys= np.std(boys_heights)
```

```python
mean_girls = np.mean(girls_heights)
std_girls= np.std(girls_heights)
```

```python
print(f"Boys ----Mean: {mean_boys}, Standard Deviation: {std_boys}")
print(f"Girls---Mean: {mean_girls}, Standard Deviation: {std_girls}")
```

```python
#Define colors for SD lines
colors = ['green', 'orange', 'red']
```

```python
plt.figure(figsize=(14,6))

#Plot for boys
plt.subplot(1,2,1)
sns.histplot(boys_heights, kde = True, stat= 'density', linewidth=0)
xmin, xmax = plt.xlim()

x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_boys, std_boys)
plt.plot(x,p,'k', linewidth=2)
title = "Boys Heights\nMean= {:.2f} , SD = {:.2f} ".format(mean_boys, std_boys)
plt.title(title)

#PLot standard deviation lines for boys

for i in range(1,4):
    plt.axvline(mean_boys + i *std_boys, color = colors[i-1], linestyle='--')
    plt.axvline(mean_boys - i *std_boys, color = colors[i-1], linestyle='--')



#Plot for girls
plt.subplot(1,2,2)
sns.histplot(girls_heights, kde = True, stat= 'density', linewidth=0)
xmin, xmax = plt.xlim()

x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_girls, std_girls)
plt.plot(x,p,'k', linewidth=2)
title = "Girls Heights\nMean= {:.2f} , SD = {:.2f} ".format(mean_girls, std_girls)
plt.title(title)

#PLot standard deviation lines for boys

for i in range(1,4):
    plt.axvline(mean_girls + i *std_girls, color = colors[i-1], linestyle='--')
    plt.axvline(mean_girls - i *std_girls, color = colors[i-1], linestyle='--')
```

```python
df_2015 = pd.read_csv('marathon_results_2015.csv')
df_2016 = pd.read_csv('marathon_results_2016.csv')
df_2017 = pd.read_csv('marathon_results_2017.csv')
```

```python
df = pd.concat([df_2015,df_2016,df_2017])
```

```python
df
```

```python
df[['hour','mins','seconds']] = df['Official Time'].str.split(':', expand = True)
```

```python
df['hour'] = df['hour'].astype(int)
df['mins'] = df['mins'].astype(int)
df['seconds'] = df['seconds'].astype(int)
```

```python
df['total_time'] = df['hour']*3600+df['mins']*60+df['seconds']
```

```python
df
```

```python
male = df[df['M/F']=='M']
female = df[df['M/F']=='F']
```

```python
male
```

```python
male_mean = male['total_time'].mean()
male_mean
male_std = male['total_time'].std()
male_std
```

```python
female_mean = female['total_time'].mean()
female_std = female['total_time'].std()
```

```python
print(f"Male ----Mean: {male_mean}, Standard Deviation: {male_std}")
print(f"Female---Mean: {female_mean}, Standard Deviation: {female_std}")
```

```python
#Define colors for SD lines
colors = ['green', 'orange', 'red']
```

```python
plt.figure(figsize=(14,6))

#Plot for males
plt.subplot(1,2,1)
sns.histplot(male['total_time'], kde = True, stat= 'density', linewidth=0)
xmin, xmax = plt.xlim()

x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x,male_mean ,male_std )
plt.plot(x,p,'k', linewidth=2)
title = "Male time\nMean= {:.2f}seconds , SD = {:.2f} ".format(male_mean,male_std )
plt.title(title)

#PLot standard deviation lines for boys

for i in range(1,4):
    plt.axvline(male_mean + i *male_std, color = colors[i-1], linestyle='--')
    plt.axvline(male_mean - i *male_std, color = colors[i-1], linestyle='--')



#Plot for Females
plt.subplot(1,2,2)
sns.histplot(female['total_time'], kde = True, stat= 'density', linewidth=0)
xmin, xmax = plt.xlim()

x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, female_mean, female_std)
plt.plot(x,p,'k', linewidth=2)
title = "Girls Heights\nMean= {:.2f}seconds, SD = {:.2f} ".format(female_mean, female_std)
plt.title(title)

#PLot standard deviation lines for boys

for i in range(1,4):
    plt.axvline(female_mean + i *female_std, color = colors[i-1], linestyle='--')
    plt.axvline(female_mean - i *female_std, color = colors[i-1], linestyle='--')

```
