```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
#generate data for zero skew
data_zero_skew= np.random.normal(loc=0,scale=1,size=1000)
#generate data for right skew
data_right_skew= np.random.exponential(scale=1,size=1000)
#generate data for left skew
data_left_skew= -np.random.exponential(scale=1,size=1000)

data_zero_skew
```

```python
mean_zero_skew=np.mean(data_zero_skew)
median_zero_skew=np.median(data_zero_skew)
mean_right_skew=np.mean(data_right_skew)
median_right_skew=np.median(data_right_skew)
mean_left_skew=np.mean(data_left_skew)
median_left_skew=np.median(data_left_skew)
```

```python
#plot histogram
plt.figure(figsize=(15,5))
```

```python
plt.subplot(1,3,1)
plt.hist(data_zero_skew,bins=30,color='blue',alpha=0.7)
plt.axvline(mean_zero_skew,color='red',linestyle='dashed',linewidth=1,label=f'Mean:{mean_zero_skew:.2f}')
plt.axvline(median_zero_skew,color='green',linestyle='dashed',linewidth=1,label=f'Median:{median_zero_skew:.2f}')
plt.title('Zero Skew')
plt.legend()
```

```python
plt.subplot(1,3,2)
plt.hist(data_right_skew,bins=30,color='orange',alpha=0.7)
plt.axvline(mean_right_skew,color='red',linestyle='dashed',linewidth=1,label=f'Mean:{mean_right_skew:.2f}')
plt.axvline(median_right_skew,color='green',linestyle='dashed',linewidth=1,label=f'Median:{median_right_skew:.2f}')
plt.title('Right Skew')
plt.legend()
```

```python
plt.subplot(1,3,3)
plt.hist(data_left_skew,bins=30,color='green',alpha=0.7)
plt.axvline(mean_left_skew,color='red',linestyle='dashed',linewidth=1,label=f'Mean:{mean_left_skew:.2f}')
plt.axvline(median_left_skew,color='green',linestyle='dashed',linewidth=1,label=f'Median:{median_left_skew:.2f}')
plt.title('Left Skew')
plt.legend()
plt.tight_layout() #used in command lines to display the representation of data
plt.show()
```
