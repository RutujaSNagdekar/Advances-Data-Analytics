```python
#For the marathon dataset find the conditional probability to finish within 3 hours given that pace is less than 8 minutes per mile/km
```

```python
import pandas as pd
```

```python
df_2015 = pd.read_csv("marathon_results_2015.csv")
df_2016 = pd.read_csv("marathon_results_2016.csv")
df_2017 = pd.read_csv("marathon_results_2017.csv")
```

```python
df = pd.concat([df_2015,df_2016,df_2017])
```

```python
df
```

```python
df[['hour','mins','seconds']] = df['Official Time'].str.split(":",expand=True)
```

```python
df
```

```python
df['hour'] = df['hour'].astype(int)
df['mins'] = df['mins'].astype(int)
df['seconds'] = df['seconds'].astype(int)
```

```python
df['total_time_in_seconds'] = df['hour']*3600+df['mins']*60+df['seconds']
```

```python
condition_A = df['total_time_in_seconds'] <=3*60*60
```

```python
df[['hour','mins','seconds']] = df['Pace'].str.split(":",expand=True)
```

```python
df['hour'] = df['hour'].astype(int)
df['mins'] = df['mins'].astype(int)
df['seconds'] = df['seconds'].astype(int)
```

```python
df['time_of_Pace'] = df['hour']*3600+df['mins']*60+df['seconds']
```

```python
condition_B = df['time_of_Pace'] <= 8*60
```

```python
P_B = len(df[condition_B])/len(df)
```

```python
P_A_and_B = len(df[condition_A & condition_B])/len(df)
```

```python
P_A_given_B = P_A_and_B/P_B
```
