```python
import pandas as pd
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('onlinefraud.csv')
```

```python
df
```

```python
df.dtypes
```

```python
features = ['type','amount','nameOrig','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']
target = 'isFraud'
```

```python
#Encoding
#Encoding 
encoder = LabelEncoder()
for col in features:
    if df[col].dtype == object:
        df[col] = encoder.fit_transform(df[col])
df
```

```python
#Scaling
scaler = StandardScaler()
df[['amount','nameOrig','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']]= scaler.fit_transform(df[['amount','nameOrig','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']])
```

```python
df
```

```python
X = df[features]
y = df[target]
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,y,test_size=0.2,random_state=1)
```

# Prediction through Naive bayes theorm

```python

nb = GaussianNB()
nb.fit(X_train,Y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
y_pred
```

```python
#Print various classification metrics
print('Confusion Matrix : ', metrics.confusion_matrix(Y_test, y_pred))
print('Accuracy: ', metrics.accuracy_score(Y_test, y_pred))
print('Precision Score: ', metrics.precision_score(Y_test, y_pred))
print('Recall  Score: ', metrics.recall_score(Y_test, y_pred))
print('F1 Score: ', metrics.f1_score(Y_test, y_pred))
print('Classification Report: ', metrics.classification_report(Y_test, y_pred))
```

# Prediction through Logistic Regression

```python
model_lr = LogisticRegression()
model_lr.fit(X_train,Y_train) 
```

```python
y_pred_lr = model_lr.predict(X_test)
```

```python
#Print various classification metrics
print('Confusion Matrix : ', metrics.confusion_matrix(Y_test, y_pred_lr))
print('Accuracy: ', metrics.accuracy_score(Y_test, y_pred_lr))
print('Precision Score: ', metrics.precision_score(Y_test, y_pred_lr))
print('Recall  Score: ', metrics.recall_score(Y_test, y_pred_lr))
print('F1 Score: ', metrics.f1_score(Y_test, y_pred_lr))
print('Classification Report: ', metrics.classification_report(Y_test, y_pred_lr))
```
