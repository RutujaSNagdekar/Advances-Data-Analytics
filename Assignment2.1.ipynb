```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('SP500 oil gold bitcoin.csv')
```

```python
#Calculate corelation
corr_overall=df[['BITCOIN','Gold','S&P500']].corr()
print(corr_overall)
```

```python
#Calculate corelation
cov_overall=df[['BITCOIN','Gold','S&P500']].cov()
print(cov_overall)
```

```python
#create scatter plots with corelation values
fig,axes=plt.subplots(1,2,figsize=(18,6))
#overall data plot for BITCOIN nad Gold
axes[0].scatter(df['BITCOIN'],df['Gold'],color='brown',alpha=0.5)
axes[0].set_title(f'Overall Correlation for BITCOIN nad Gold:{corr_overall.loc['BITCOIN','Gold']:.2f}')
axes[0].set_xlabel('BITCOIN')
axes[0].set_ylabel('Gold')
axes[0].grid(True)

#overall data plot for Bitcoin and s&p500
axes[1].scatter(df['BITCOIN'],df['S&P500'],color='black',alpha=0.5)
axes[1].set_title(f'Overall Correlation for BITCOIN and S&P500:{corr_overall.loc['BITCOIN','S&P500']:.2f}')
axes[1].set_xlabel('BITCOIN')
axes[1].set_ylabel('S&P500')
axes[1].grid(True)

#create scatter plots with covariance values
fig,axes=plt.subplots(1,2,figsize=(18,6))
#overall data plot for BITCOIN nad Gold
axes[0].scatter(df['BITCOIN'],df['Gold'],color='purple',alpha=0.5)
axes[0].set_title(f'Overall Covariance for BITCOIN nad Gold:{cov_overall.loc['BITCOIN','Gold']:.2f}')
axes[0].set_xlabel('BITCOIN')
axes[0].set_ylabel('Gold')
axes[0].grid(True)

#overall data plot for Bitcoin and s&p500
axes[1].scatter(df['BITCOIN'],df['S&P500'],color='pink',alpha=0.5)
axes[1].set_title(f'Overall Covariance for BITCOIN and S&P500:{cov_overall.loc['BITCOIN','S&P500']:.2f}')
axes[1].set_xlabel('BITCOIN')
axes[1].set_ylabel('S&P500')
axes[1].grid(True)


```

```python
#heatmap for Corealtion
import seaborn as sns
plt.figure(figsize=(6,4))
sns.heatmap(corr_overall,annot=True,fmt=".2f",cmap="coolwarm")
plt.title("overall correlation heat map")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()

#heatmap for Covariance
plt.figure(figsize=(6,4))
sns.heatmap(cov_overall,annot=True,fmt=".2f",cmap="Paired")
plt.title("overall covariance heat map")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
