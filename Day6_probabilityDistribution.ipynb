# Uniform Distribution

```python
#probability distributions: Discrete and Uniform Distribution

import matplotlib.pyplot as plt
import numpy as np
```

```python
#Number of dice rolls
num_rolls = 6
```

```python
#Calculate probabilities for each outcomes(1 to 6)
probabilities = np.ones(6)/6
probabilities
```

```python
#Calculate the cdf (cumulative distribution function)
cdf = np.cumsum(probabilities)
cdf
```

```python
#Plot the PMF and CDF 
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.bar(range(1,7),probabilities)
plt.xlabel("Dice Rolls")
plt.ylabel("Probabilities")
plt.title("PMF of dice roll")

plt.subplot(1,2,2)
plt.bar(range(1,7),cdf,color='green')
plt.xlabel("Dice Rolls")
plt.ylabel(" Cumulative Probabilities")
plt.title("CDF of dice roll")
plt.tight_layout()
```

```python
# Practical example of PMF and CDF in binomila distribution
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
tips_df = pd.read_csv("Tips.csv")
```

```python
#Binary outcome Success if (tip/Total bill) > 0.15 otherwise Failure
```

```python
def classify_tip(row):
    total_bill = row['total_bill']
    tip= row['tip']
    if (tip/total_bill) > 0.15:
        return 1 #Success
    else:
        return 0 #Failure

```

```python
#Apply the function to create the binary column
tips_df['tip_binary'] = tips_df.apply(classify_tip, axis=1)
```

```python
#Total trials
n = len(tips_df)
n
```

```python
#Number of success 
k= tips_df['tip_binary'].sum()
print(k)
```

```python
#Probability of success
p= k/n
p
```

```python

#Define number of trials for binary distributiuon 
trials = 100
hypothetical_k= 50
```

```python
#Binomial probability of exacly k success in trials
probability = binom.pmf(hypothetical_k, trials, p)
probability
```

```python
#Exact binomial probability for 60% success (60 out of 100 trials)

min_desired_successes = int(0.6*trials)
exact_probability_60 = binom.pmf(min_desired_successes,trials,p)
print(exact_probability_60)

#Just to calculate 65% success
exact_probability_65 = binom.pmf(65,trials,p)
print(exact_probability_65)
```

```python
cumulative_prob = binom.cdf(min_desired_successes,trials,p)
cumulative_prob
```

```python
#Output the result 
print(f"Total trials(n): {n}")
print(f"Number of success: {k}")
print(f"Probability  of success: {p:.4f}")
print(f"Binomial probability of exactly {hypothetical_k} in {trials} is { probability: .4f}")
print(f"Binomial probability at 60% success in {trials} is { exact_probability_60: .4f}")
print(f"Binomial probability at 65% success in {trials} is { exact_probability_65: .4f}")
print(f"Cumulative probability at 60% or fewer in {trials} is { cumulative_prob: .4f}")
```

```python
#Generate x values (number of successes from 0 to 100)
x = np.arange(0,trials+1)
x
```

```python
#Calculate the PMF for each numnber of successes
pmf_values = binom.pmf(x,trials,p)
pmf_values
```

```python
#CalculATE  rhe cdf for each number of success
cdf_values = binom.cdf(x,trials,p)
cdf_values
```

```python
#Plotting the pmf and cdf
plt.figure(figsize=(12,6))
#PMF plot
plt.subplot(1,2,1)
plt.bar(x,pmf_values,color='skyblue',alpha=0.7)
plt.title(f"Binomial Distribution PMF (n={trials},p={p:.4f})")
plt.xlabel("number of Successes")
plt.ylabel("probability")
plt.xticks(np.arange(0,trials+1,5))
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.axvline(x=min_desired_successes,color='red',linestyle='--',label='60 Successes')
plt.axvline(x=65,color='green',linestyle='--',label='Exact 65 Successes')
plt.legend()
```

```python
#Plotting the pmf and cdf
plt.figure(figsize=(10,6))
#PMF plot
plt.subplot(1,2,1)
plt.bar(x,pmf_values,color='skyblue',alpha=0.7)
plt.title(f"Binomial Distribution PMF (n={trials},p={p:.4f})")
plt.xlabel("number of Successes")
plt.ylabel("probability")
plt.xticks(np.arange(0,trials+1,5))
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.axvline(x=min_desired_successes,color='red',linestyle='--',label='60 Successes')
plt.axvline(x=65,color='green',linestyle='--',label='Exact 65 Successes')
plt.legend()



#CDF Plot
plt.figure(figsize=(10,6))
#PMF plot
plt.subplot(1,2,2)
plt.plot(x,cdf_values,color='orange',label='CDF',markersize=4,marker='o')
plt.title(f"Binomial Distribution CDF (n={trials},p={p:.4f})")
plt.xlabel("number of Successes")
plt.ylabel("Cumulative probability")
plt.xticks(np.arange(0,trials+1,5))
plt.grid()
plt.axhline(y= cumulative_prob,color='blue',linestyle='--',label='Cumulative Probability at 60')
plt.axvline(x=min_desired_successes,color='red',linestyle='--',label='60 Successes')
plt.legend()

plt.tight_layout()
```
