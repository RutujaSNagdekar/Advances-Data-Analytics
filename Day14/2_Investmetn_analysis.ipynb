```python
#inverstment analysis
import numpy as np
import matplotlib.pyplot as plt
```

```python
initial_investment = 10000  #Initial invertment in dollars
mean_return = 0.15   #Expected annual returns
std_dev = 0.15     # Std devialtions of returns 
num_simulations = 1000    #Number of monte carlo simulations    
```

```python
#Monte carlo simulations 
simulated_end_values = []
for _  in  range(num_simulations):
    #Generated random annual return for each simulation 
    random_return = np.random.normal(mean_return, std_dev)
    print(random_return)
    #End value at the end of one year of investment 
    end_value = initial_investment * (1 + random_return)

    #store it in the list
    simulated_end_values.append(end_value)
```

```python
print(simulated_end_values)
```

```python
#Results 
mean_end_value = np.mean(simulated_end_values) #MEan of all simulated end values 
mean_end_value
```

```python
#Proportions of cases where we have loss of (End annual value < Initial investment )
```

```python
risk_of_loss = np.mean(np.array(simulated_end_values) < initial_investment) 
```

```python
risk_of_loss
```
