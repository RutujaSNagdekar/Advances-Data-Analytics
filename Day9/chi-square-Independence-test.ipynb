```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv('airline_passenger_satisfaction.csv')
df
```

```python
male_df = df[df['Gender'] == 'Male']
category1 = 'Class'
category2 = 'On-board Service'
```

```python
crosstab = pd.crosstab(male_df[category1],male_df[category2])
```

```python
crosstab
```

```python
#perform chi-square test
chi2,pval,degree_of_freedom,expected_counts = chi2_contingency(crosstab)
expected_counts
```

```python
#print results
print(f"\nCategory 1: {category1}, Category 2: {category2} (Male Passengers)")
print(crosstab)
print(f"CHi-square: {chi2:.2f},p-value: {pval:.4f} , Degree of Freedom : {degree_of_freedom}")
```

```python
# Interpretation
if pval < 0.05:
    print("Reject null hypothesis: Thers's significance association between class and on board service ")
else:
    print("fail to reject null hypothesis, Thers's  no significance association class and on board service ")
```
