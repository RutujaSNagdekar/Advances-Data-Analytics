```python
import pandas as pd
```

```python
df_1 = pd.read_csv('Men ODI Player Innings Stats - 20th Century.csv')
df_2 = pd.read_csv('Men ODI Player Innings Stats - 21st Century.csv')
```

```python
df = pd.concat([df_1,df_2])
```

```python
df_country= df[(df['Country'] == 'India') | (df['Country'] == 'Australia')]
```

```python
#Find the list of bowlers who have taken at least 25 wickets overall(sum).
#check if your results are correct by comparing with sites such as CricInfo
#(.g wickets taken by the bowler who retired before 2020)
```

```python
df_country['Innings Wickets Taken'] = df_country['Innings Wickets Taken'].replace('-' , 0)
df_country['Innings Wickets Taken'] = df_country['Innings Wickets Taken'].replace('Nan' , 0)
df_country['Innings Wickets Taken'] = df_country['Innings Wickets Taken'].replace('' , 0)
df_country['Innings Wickets Taken'] = df_country['Innings Wickets Taken'].astype(int)

```

```python
df_country
```

```python
df_group = df_country.groupby('Innings Player')
```

```python
total_wickets = df_group['Innings Wickets Taken'].sum()
```

```python
total_wickets = total_wickets[total_wickets > 25]
```

```python
df_country = pd.merge(df_country,total_wickets , on = 'Innings Player' , how = 'left')
```
